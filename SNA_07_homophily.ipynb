```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def create_graph():
    G=nx.Graph()
    G.add_nodes_from(range(20))
    return G
```

```python
def assign_gpa(G):
    for each in G.nodes():
        G.nodes[each]['name']=random.randint(0,10)
        G.nodes[each]['type']='person'
```

```python
def get_labels(G):
    dict1={}
    for each in G.nodes():
        dict1[each]=G.nodes[each]['name']
    return dict1
        
```

```python
def visualize(G,dictlabels):
    nx.circular_layout(G)
    nx.draw(G,labels=dictlabels)
    plt.show()
```

```python
G=create_graph()
assign_gpa(G)
dictlabels=get_labels(G)
visualize(G,dictlabels)
```

```python
def get_size(G):
    array1=[]
    for each in G.nodes():
        array1.append(G.nodes[each]['name'])
    return array1
```

```python
def visualize(G,labeldict,nodesize):
    nx.circular_layout(G)
    nx.draw(G,labels=labeldict,node_size=nodesize)
    plt.show()
```

```python
nodesize=get_size(G)
print(nodesize)
visualize(G,dictlabels,[x*20 for x in nodesize])
```

```python
def add_foci_nodes(G):
    n=G.number_of_nodes()
    i=n+1
    foci_nodes=['gym','eatout','movie_club','karate_club','yoga_club']
    for j in range(0,5):
        G.add_node(i)
        G.nodes[i]['name']=foci_nodes[j]
        G.nodes[i]['type']='foci'
        i=i+1
```

```python
def get_size(G):
    array2=[]
    for each in G.nodes():
        if G.nodes[each]['type']=='person':
            array2.append(G.nodes[each]['name']*200)
        else:
            array2.append(100)
    return array2
```

```python
def get_colors(G):
    c=[]
    for each in G.nodes():
        if G.nodes[each]['type']=='person':
            c.append('green')
        else:
            c.append('yellow')
    return c
```

```python
def visualize(G,labeldict,nodesize,color):
    nx.circular_layout(G)
    nx.draw(G,labels=labeldict,node_size=nodesize,node_color=color)
    plt.show()
```

```python
G=create_graph()
assign_gpa(G)
add_foci_nodes(G)
labeldict=get_labels(G)
nodesize=get_size(G)
color_array=get_colors(G)
visualize(G,labeldict,nodesize,color_array)
```

```python
def get_foci_nodes(G):
    f=[]
    for each in G.nodes():
        if G.nodes[each]['type']=='foci':
            f.append(each)
    return f
get_foci_nodes(G)
```

```python
def get_person_nodes(G):
    p=[]
    for each in G.nodes():
        if G.nodes[each]['type']=='person':
            p.append(each)
    return p
```

```python
def add_foci_edges(G):
    foci_nodes=get_foci_nodes(G)
    person_nodes=get_person_nodes(G)
    for each in person_nodes:
        r=random.choice(foci_nodes)
        G.add_edge(each,r)
    return G
```

```python
G=add_foci_edges(G)
visualize(G,labeldict,nodesize,color_array)
```
