```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_edge(0,1)
G.add_edge(0,6)
G.add_edge(5,4)
G.add_edge(5,6)
G.add_edge(6,3)
G.add_edge(2,3)
G.add_edge(1,6)
G.add_edge(2,3)
G.add_edge(6,2)
G.add_edge(6,4)

nx.draw(G, with_labels = 1)
```

```python
# Finding the betweenness centrality for node "0"
nodes = G.nodes()
print("Nodes in graph", nodes)
len1 = len(nodes)
print("Number of nodes", len1)

for n in nodes:
    print("\n")
    print('\033[94m'+'Betweeness centrality for node '+'\033[94m',n)
    betweeness_centrality = 0
    count=0
    print("\n")

    for i in range(0, len1):
        for j in range(i+1, len1):
            sp = list(nx.all_shortest_paths(G, i, j))
            print('1.','(', i,",", j,')'," All shortest paths: ", sp,"and",'no.of paths :',len(sp))


            count = 0
            for p in sp:
                print("p :", p)
                between = len(p)-2
                if between >=1:
                    print("2. Path between",i,"and ",j,"is", p)
                    print("3. Number of between nodes", between)
                    x=1

                    while x<=between:
                        if p[x]==n:
                            count+=1
                        x+=1
            k=len(nx.shortest_path(G, i, j))-2

            if count>=1 and k>=1:
                print("4. Number of shortest paths between",i,"and ", j, "is", k)
                print("5. Path through ",n,'is: ', count)
                betweeness_centrality+=count/len(sp)
                print("\n")
            else:
                print("\n")


    print("\n Node",n,"centrality :", betweeness_centrality)
    print("\n")
```
