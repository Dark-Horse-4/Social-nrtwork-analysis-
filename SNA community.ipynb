```python
G=nx.barbell_graph(4,0)
```

```python
nx.draw(G,with_labels=True)
```

```python
G.edges()
```

```python
G.remove_edge(0,3)
G.remove_edge(1,2)
nx.draw(G,with_labels=True)
```

```python
first_community=[]
comb=[list(x) for x in itertools.combinations(G.nodes(),4)]
print("comb",comb)
first_community.extend(comb)
print("\n First community",first_community)

second_community=[]
for i in range(len(first_community)):
    l=list(set(G.nodes())-set(first_community[i]))
    second_community.append(l)
print("\n second community",second_community)
```
