```python
import networkx as nx
G = nx.Graph()
G.add_edges_from([(3,1),(3,2)])
nx.draw_networkx(G, with_labels = True, node_color ='blue')
```

```python
if G.has_edge(1,2):
    print('yes')
else:
    print('no')

if G.has_edge(1,3) and G.has_edge(2,3):
    print('common node availble: 3')
    G.add_edge(1,2)
    nx.draw(G,with_labels = True)
else:
    print('no common nodes available')
```
