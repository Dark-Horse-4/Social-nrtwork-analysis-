```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_edge(1,4)
G.add_edge(1,2)
G.add_edge(6,5)
G.add_edge(4,5)
G.add_edge(2,4)
G.add_edge(4,6)
G.add_edge(3,5)
G.add_edge(1,3)

nx.draw(G, with_labels = 1)
```

```python
sp   = nx.shortest_path(G,1,4)
print(sp)
print(len(sp))
```

```python
sp = nx.shortest_path(G,1,5)
print(sp)
print(len(sp))
```

```python
def closeness_centrality(G,i):
    sum = 0
    centratlity = 0
    lenght = len(G.nodes())
    for j in range (0, lenght):
        sp = nx.shortest_path(G,j+1,i)
        sum += len(sp)-1
    centrality= sum/(lenght-1)
    return centrality
```

```python
print("Nodes\tCloseness centrality\n")
for i in range (0, len(G.nodes())):
    c= closeness_centrality(G, i+1)
    print(i+1, "\t",c,'\n')
```
