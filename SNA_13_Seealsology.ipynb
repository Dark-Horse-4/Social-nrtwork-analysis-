```python
train_edges = train_edges[0:50]
```

```python
def get_graph(x):
    G = nx.DiGraph()
    G=nx.from_pandas_edgelist(x,'source', 'target', create_using=nx.DiGraph())
    return G
```

```python
G = get_graph(train_edges)
nx.draw(G,with_labels =True)
```
