```python
import networkx as nx
G = nx.Graph()
G.add_edges_from([(3,1),(3,2),(1,5),(2,5),(3,4)])
nx.draw_networkx(G, with_labels = True, node_color ='blue')
```

```python
nb = sorted(nx.common_neighbors(G,1,2))
print(nb)
nb = sorted(nx.common_neighbors(G,4,5))
print(nb)
print(len(nb))
```

```python
for u in G.nodes():
    for v in G.nodes():
        if u != v:
            if len(sorted(nx.common_neighbors(G,u,v)))>0:   
                if G.has_edge(u,v)==False:
                    print('u',u,'v',v,'common neighbors',sorted(nx.common_neighbors(G,u,v)),'edge between',u,'and',v,'created')
                    G.add_edge(u,v)
nx.draw(G,with_labels = True)
             
```
