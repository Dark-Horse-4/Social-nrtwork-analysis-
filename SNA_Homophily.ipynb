```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
def create_graph():
    G=nx.Graph()
    for i in range(1,20):
        G.add_node(i)
    return G
```

```python
def assign_gpa(G):
    for each in G.nodes():
        G.nodes[each]['name']=random.randint(0,10)
        G.nodes[each]['type']='person'
```

```python
def get_labels(G):
    dict1={}
    for each in G.nodes():
        dict1[each]=G.nodes[each]['name']
    return dict1    
```

```python
def visualize(G,labeldict):
    nx.circular_layout(G)
    nx.draw(G,labels=labeldict)
    plt.show()
```

```python
G=create_graph()
assign_gpa(G)
labeldict=get_labels(G)
visualize(G,labeldict)
```

```python
def get_size(G):
    array1=[]
    for each in G.nodes():
        array1.append(G.nodes[each]['name']*20)
    return array1   
```

```python
def visualize(G,labeldict,nodesize):
    nx.circular_layout(G)
    nx.draw(G,labels=labeldict,node_size= nodesize)
    plt.show()
```

```python
nodesize=get_size(G)
print(nodesize)
visualize(G,labeldict,nodesize)
```

```python
def get_colors(G):
    c= []
    for each in G.nodes():
        if G.nodes[each]['type']  == 'person':
            c.append('green')
        else:
            c.append('yellow')
    return c
```
