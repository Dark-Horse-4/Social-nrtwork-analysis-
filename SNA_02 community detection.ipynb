# Finding the communities in the Network

```python
import networkx as nx

import matplotlib.pyplot as plt
import itertools
```

```python
# for i in itertools.combinations([1,2,3,4,5,6,7],2):
#     print(list(i))
```

```python
G=nx.barbell_graph(5,0)
```

```python
nx.draw(G,with_labels=1)
```

```python
G.edges()
G.remove_edge(1,3)

G.remove_edge(7,5)
nx.draw(G,with_labels=1)
```

```python
first_community=[]
comb = [list(x) for x in itertools.combinations(G.nodes(),4)]
print("comb ",comb)
first_community.extend(comb)
print("\nFirst Community",first_community)

#print(G.nodes())
#print(set(first_community[0]))

second_community=[]
for i in range(len(first_community)):
    l = list(set(G.nodes())-set(first_community[i]))
    second_community.append(l)
print("\nSecond Community",second_community)
```

```python
nx.draw(G,with_labels=1)
```

```python
#which division is best
num_intra_edges1=[]
num_intra_edges2=[]
num_inter_edges=[]
ratio=[]
for i in range(len(first_community)):
    num_intra_edges1.append(G.subgraph(first_community[i]).number_of_edges())
    print(first_community[i]," ",G.subgraph(first_community[i]).number_of_edges())
print("Community 1 No. of intra edges:",num_intra_edges1)
```

```python
for i in range(len(second_community)):
    num_intra_edges2.append(G.subgraph(second_community[i]).number_of_edges())
    print(second_community[i]," ",G.subgraph(second_community[i]).number_of_edges())
print("Community 1 No. of intra edges:",num_intra_edges2)
```

```python
e=G.number_of_edges()
for i in range(len(first_community)):
    num_inter_edges.append(e-num_intra_edges1[i]-num_intra_edges2[i])
    print("e",e," ,community1 edges",num_intra_edges1[i]," ,community2 edges",num_intra_edges1[i]," ,inter edges",e-num_intra_edges1[i]-num_intra_edges2[i])
```

```python
#Find the ratio
print(len(first_community))
print(len(second_community))

for i in range(len(first_community)):
    ratio.append((float)(num_intra_edges1[i])/num_inter_edges[i])
max_value=max(ratio)
max_index=ratio.index(max_value)
print(max_index)
print("(", first_community[max_index],")(", second_community[max_index],")")
nx.draw(G,with_labels=1)
```
