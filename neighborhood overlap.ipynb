```python
import networkx as nx
G = nx.Graph()
G.add_edges_from([('A',1),('A',2),('A',3),('A',4),('A',5),('B',6),('B',7),('B',8),
                ('A',9),('A',10),('A',11),('A',12),('A',13),('B',9),('B',10),('B',11),('B',12),('B',13)])
nx.draw_networkx(G, with_labels = True, node_color ='blue')
```

```python
def neigh_ovelap(G,n1,n2):
    A_neighbors = []
    for n in G.neighbors(n1):
        A_neighbors.append(n)
    
    B_neighbors = []
    for n in G.neighbors(n2):
        B_neighbors.append(n)
    
    overlap_nodes =[]

    for a in A_neighbors:
            if a in B_neighbors:
                overlap_nodes.append(a)  
   
    neighborhood_overlap = len(overlap_nodes)/(len(A_neighbors)+len(B_neighbors)-len(overlap_nodes))
    
    print('the overlapping nodes are :',overlap_nodes)
    print('The neighborhood overlap is :')
    
    return neighborhood_overlap
```

```python
neigh_ovelap(G,'A','B')
```
